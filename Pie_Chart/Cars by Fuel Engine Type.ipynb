#### Importing the necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### Importing the dataset

```python
df_car_engine = pd.read_csv("cars_by_fuel_engine_type_data.csv")
df_car_engine
```

#### Creating the Pie Chart

```python
# increasing the size/ dimensions of the figure 
plt.figure(figsize = (10, 8))   

# defining bar charts attributes
plt.pie( df_car_engine["Number of Cars"],
         labels = df_car_engine["Engine Fuel Type"].values,
         autopct = '%.2f%%',
         textprops = {'size' : 'x-large',
                      'fontweight' : 'bold',
                      'rotation' : '30',
                      'color':'w'})
       


# coloring the pie chart 
sns.set_palette('colorblind')

# adding a title for the bar chart
plt.title("Cars by Fuel Engine Type", fontsize = 16, fontweight = 'bold')


# displaying the final bar chart
plt.legend()
plt.show()

# saving the figure
plt.savefig("Cars by Fuel Engine Type")
```
